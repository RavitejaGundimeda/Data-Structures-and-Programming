```python
import matplotlib.pyplot as plt
import math
import random
import time
import numpy as np 

def median_of_three(arr, p, r):
  mid = (p+r-1)//2
  a = arr[p]
  b = arr[mid]
  c = arr[r-1]
  if a <= b and b <= c:
    return b, mid
  if c <= b and b <= a:
    return b, mid
  if a <= c and c <= b:
    return c, r-1
  if b <= c and c <= a:
    return c, r-1
  return a, p

def plott(x,y,title):
    plt.plot(x, y)
    # naming the x axis
    plt.xlabel('Input-N')
   # naming the y axis
    plt.ylabel('Time in seconds')
  
   # giving a title to my graph
    plt.title(title)
  
  # function to show the plot
    plt.show()

def partition(arr,p,r):
  piv, ind = median_of_three(arr,p,r)  #taking median element
  arr[p], arr[ind] = arr[ind], arr[p]
  i = p
  for j in range(p+1,r+1):
    
    if(arr[j] <= piv):
      i+=1
      arr[i],arr[j] = arr[j],arr[i]
      
  arr[p],arr[i] = arr[i],arr[p] #keeping pivot at the perfect position
  return i



def quicksort(arr,p,r):
  if p < r:
    q = partition(arr, p, r)
    quicksort(arr, p, q-1)
    quicksort(arr, q+1, r)

#arr = [10, 80, 30, 90, 40, 50, 70]
#print(arr)
#quicksort(arr, 0, len(arr)-1)
#print(arr)
time_list =[]
for n in range(10,500,10):
  arr=[]
  arr = np.random.randint(1,1000,n)
  arr.sort() #For worst case input
  #print("Initial array is ",arr)
  #print("length of array is ",len(arr))
  #print(A)
  start,end =0.0,0.0
  start = time.time()
  quicksort(arr, 0, len(arr)-1)
  end = time.time()
  time_res = end - start
  #print(time_res)
  time_list.append(time_res)
  #print("result is ", arr)

x_axis = [i for i in range(10,500,10)]
plott(x_axis,time_list,"Quick_Sort_Median")
```
