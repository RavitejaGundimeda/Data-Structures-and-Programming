Maximal sub-array problem - Brute Force approach

```python
import matplotlib.pyplot as plt
import math
import random
import time

def plott(x,y,title):
    plt.plot(x, y)
    # naming the x axis
    plt.xlabel('Input-N')
   # naming the y axis
    plt.ylabel('Time in seconds')
  
   # giving a title to my graph
    plt.title(title)
  
  # function to show the plot
    plt.show()

def Cross_Sum(A,low,mid,high):
  sum = 0
  max_left = -9999
  for i in range(mid, low-1, -1):
        sum = sum + A[i]
 
        if (sum > max_left):
            max_left = sum
 
    
  sum = 0
  max_right = -9999
  for i in range(mid + 1, high + 1):
    sum = sum + A[i] 
    if (sum > max_right):
      max_right = sum
 
    
  return max(max_left + max_right, max_left, max_right)
 



def MSA_DC(A,low,high):
  #divide and conquer approach
  if low == high:
    return A[low]
  else:
    mid = (low + high)//2
    return max(MSA_DC(A,low,mid),MSA_DC(A,mid+1,high),Cross_Sum(A,low,mid,high))


time_list =[]
for n in range(10,500,5):
  A=[]
  start,end =0.0,0.0
  A = random.sample(range(-n,n),n) 
  #print(A)
  start = time.time()
  max_sum = MSA_DC(A,0,len(A)-1)
  #print(max_sum)
  end = time.time()
  time_res = end - start
  #print(time_res)
  time_list.append(time_res)


x_axis = [i for i in range(10,500,5)]
plott(x_axis,time_list,"MSA_Divide & Conquer_nlogn")
```
